<h1>CREATING A CLASSIFIER FROM LOAN DATA</h1>

```python
#import pandas and see what data looks like
import pandas as pd

data = pd.read_csv('loan_train.csv')
data


```

First we need to clean up the data.

Several columns have either yes or no as values.
I'm going to change those to either 1 or 0

```python
##clean up gender column
data['Male'] =  data['Gender'].apply(lambda x: 1 if x == 'Male' else 0)
data  = data.drop('Gender', axis=1) 
data
```

```python
##clean up education column
data['Graduate'] =  data['Education'].apply(lambda x: 1 if x == 'Graduate' else 0)
data  = data.drop('Education', axis=1) 
data
```

```python
#clean up married and self employed columns
data['Is_Married'] =  data['Married'].apply(lambda x: 1 if x == 'Yes' else 0)
data  = data.drop('Married', axis=1) 
data

data['Is_SelfEmployed'] =  data['Self_Employed'].apply(lambda x: 1 if x == 'Yes' else 0)
data  = data.drop('Self_Employed', axis=1) 
data
```

Now we have to one-hot encode the property_Area and dependents columns


```python
propertyDummies = pd.get_dummies(data['Property_Area'])


data['PropertyArea_Rural'] = propertyDummies['Rural']
data['PropertyArea_Semiurban'] = propertyDummies['Semiurban']
data['PropertyArea_Urban'] = propertyDummies['Urban']
data = data.drop('Property_Area', axis=1)



data = data.dropna()
data = data.set_index('Loan_ID')
data
```

```python
depDummies = pd.get_dummies(data['Dependents'])
data['Dependents_0'] = depDummies['0']
data['Dependents_1'] = depDummies['1']
data['Dependents_2'] = depDummies['2']
data['Dependents_3+'] = depDummies['3+']
data = data.drop('Dependents', axis=1)
data
```

Now the data is in a format we can use.
All category columns have been encoding in some way.

Next on to creating a classifier, but first...

<h1>Approach to testing and classification</h1>

Since our test data doesn't have labels we won't be be able to use it for tuning the classifer. We're going to have to split the training data up.

After splitting up the data we are going to need to think about different classifiers and which might be the best to use.

```python
from sklearn.model_selection import train_test_split


featureColumns = data.columns.values.tolist()
featureColumns.remove('Loan_Status')
labelColumn = 'Loan_Status'

loans_train, loans_test = train_test_split(data, test_size=.20)

loans_train_features = loans_train[featureColumns]
loans_train_labels = loans_train[labelColumn]

loans_test_features = loans_test[featureColumns]
loans_test_labels = loans_test[labelColumn]

```

<h1>Lets see how different classifiers work</h1>

<h3>Decision Tree</h3>

```python
##lets see how a decision tree does
from sklearn import tree
from sklearn.metrics import accuracy_score

scores = pd.Series()
for i in range(100):
    treeclf = tree.DecisionTreeClassifier(criterion='entropy')
    treeclf.fit(loans_train_features, loans_train_labels)

    predictions = treeclf.predict(loans_test_features)

    scores.loc[len(scores)] = accuracy_score(predictions, loans_test_labels)

scores.describe()
```

<h3>Bagging Classifer with decision tree</h3>


```python
##lets use the same parameters as above and run 100 times

from sklearn.ensemble import BaggingClassifier

clf = tree.DecisionTreeClassifier(criterion='entropy')
bagging_clf = BaggingClassifier(clf, n_estimators=20, max_samples=.7, max_features=.7,bootstrap=True, n_jobs=-1)
bagging_clf.fit(loans_train_features, loans_train_labels)
predictions = bagging_clf.predict(loans_test_features)
accuracy_score(predictions, loans_test_labels)



```

So running the individual decision tree and the bagging classifier yielded highly variable results. Bagging always had better results, but the difference was also highly variable.
