# <h1><font color='red'>OTTO GROUP PROJECT</font></h1>


The Otto Group is an e-commerce company that sells millions of products, with many more added to their line frequently.

Their issue is that due to coordination and infrastructure challenges, many identical products are classified and tracked differently. This causes problems when analzying sales and product perfomance. 

Their request is to build a model that accurately groups products into one of several main product categories.





# DATA

The first step as always is too take a look at the data.

We are told each row corresponds to a single product with 93 numerical features each.
Each feature represents counts of different events.

```python
import pandas as pd
import numpy as np

otto_train = pd.read_csv("ottotrain.csv")
otto_train = otto_train.set_index("id")
```

Lets see how many of each target there are

# Baseline testing

Since we already have the data in a usable format, all we need to do is perform train test splitting in order to test out using a basic decision tree.

```python
from sklearn.model_selection import train_test_split

featurecols = otto_train.columns.tolist()
featurecols.remove('target')


train_features, test_features, train_labels, test_labels = train_test_split(otto_train[featurecols], otto_train['target'], test_size=.2)
featurecols
```

Now lets create the classifier, train it and test

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

clf = DecisionTreeClassifier(criterion='entropy')

clf.fit(train_features, train_labels)
predictions = clf.predict(test_features)

accuracy_score(test_labels, predictions)
```

A basic decision tree is between 70 and 71% accurate

# Neural Network

The basic thought process with a neural network is the following.

We are going to be created a fully connected NN with one hidden layer.
The first layer will 93 inputs (one for each feature). 

The hidden layer will have 40 nodes, and the last layer will have 9 nodes, one for each class of product.

We'll need to convert the dataframe into a numpy array

Let's import the necessary packages

```python
import keras
from keras import models
from keras import layers
```

```python
network_train_features = np.array(train_features)
network_test_features = np.array(test_features)

network_train_features[0]
```

For the labels lets keep only the number part of the text and convert it to an integer

```python
train_labels
```

```python
##creating this function that will transform our labels into desired format
def convertlabels(labels):
  templabels = []
  for label in labels:
    newlabel = int(label[6:])
    templabels.append(newlabel)
  labels = np.array(templabels)
  return labels
```

```python
network_train_labels = convertlabels(train_labels)
network_test_labels = convertlabels(test_labels)
```

```python
network_train_labels
```

```python
input_size = 93
hidden_size = 40
output_size = 10



network = models.Sequential()
network.add(layers.Dense(hidden_size, activation='relu', input_shape=(input_size,)))
network.add(layers.Dense(output_size, activation='softmax'))
```

```python
network.compile(optimizer='rmsprop', loss='SparseCategoricalCrossentropy', metrics=['accuracy'])
```

```python
network.fit(network_train_features, network_train_labels, epochs=10, batch_size=128)
```

Now lets test test the accuracy on the test data

```python
test_loss, test_acc = network.evaluate(network_test_features, network_test_labels)
print('test_acc:', test_acc)
```

Given the above accuracy of 78%, this represents an almost 10% improvement on the decision tree of 71%

# Adjusting the network

Now that we know a network works well for this problem, lets make some adjustments.

```python
input_size = 93
hidden = [60, 30]
output_size = 10



network = models.Sequential()
network.add(layers.Dense(hidden[0], activation='relu', input_shape=(input_size,)))
network.add(layers.Dense(hidden[1], activation='sigmoid'))
network.add(layers.Dense(output_size, activation='softmax'))
```

```python
network.compile(optimizer='rmsprop', loss='SparseCategoricalCrossentropy', metrics=['accuracy'])
```

```python
network.fit(network_train_features, network_train_labels, epochs=10)
```

```python
test_loss, test_acc = network.evaluate(network_test_features, network_test_labels)
print('test_acc:', test_acc)
```

# XGBOOST

Another option to use is XGBoost. 
An error I made in the previous sections was declaring our data as "unstructured". After some research unstructured essentially means that features don't have any visible structure to humans, or that each feature could be all of the same unit, i.e pixels in an image, sounds in an audio file.
Although the the meaning of feature is unknown to us (those using the kaggle dataset), they do have a meaning. 
This is called Tabular Data and XGBoost is highly performant when it comes to this type of data.


```python
otto_train
```

```python
#import libraries and encode features in a form that the model can use

from xgboost import XGBClassifier
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore')
ottosparse = enc.fit_transform(otto_train[featurecols])
ottosparse 
```

```python
xg_train_features, xg_test_features, xg_train_labels, xg_test_labels = train_test_split(ottosparse, otto_train['target'], test_size=.2)
xg_train_features
```

```python
params = {'tree_method':'gpu_hist', 'predictor':'gpu_predictor' }

model = XGBClassifier(params)
```

We are going to be performing randomized grid search on the following parameters to find the best performing combination..

```python
param_grid = {
    'n_estimators': [50, 100, 150, 200, 300, 400],
    'max_depth': [2, 4, 6, 8, 10, 12, 15],
    'learning_rate': [0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ]
}
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import RandomizedSearchCV
param_comb = 5
folds=5
skf = StratifiedKFold(n_splits=folds, shuffle = True, random_state = 1001)
random_search = RandomizedSearchCV(model, param_distributions=param_grid, n_iter=param_comb,  n_jobs=-1, 
                                   cv=skf.split(xg_train_features,xg_train_labels), verbose=3)
```

```python
grid_result = random_search.fit(xg_train_features,xg_train_labels)
```

```python
random_search.best_params_
```

```python
predictions = random_search.best_estimator_.predict(xg_test_features)
accuracy_score(xg_test_labels, predictions)
```

With xgboost and some hyper parameter tuning we were able to get to 80%.
Much better than the 71% of the decision tree and even the 78% of the neural network.

Let's verify that these hyper parameters give us the results we expect.

```python
tunedXGB = XGBClassifier(n_estimators=200, predictor='gpu_predictor', tree_method='gpu_hist', max_depth=10, learning_rate=.15)
```

```python
tunedXGB.fit(xg_train_features, xg_train_labels)
```

```python
tunedPredictions = tunedXGB.predict(xg_test_features)
accuracy_score(xg_test_labels, tunedPredictions)
```
